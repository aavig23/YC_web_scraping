```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#Use of requests libr: 
headers= {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36'
} #User agent_address
response=requests.get('https://www.ycombinator.com/companies/industry/health-tech',headers=headers)
webpage=response.text
print(webpage)

```
